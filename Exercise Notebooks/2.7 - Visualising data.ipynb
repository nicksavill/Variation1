Conflict between humans and tigers threatens tiger populations, kills people, and reduces public support for conservation. 

Researchers investigated causes of human deaths by tigers near the protected area of Chitwan National Park, Nepal. Eighty-eight people were killed by 36 individual tigers between 1979 and 2006, mainly within 1 km of the park edge. Such information may be helpful to identify activities that increase vulnerability to attack.

The file [Datasets/tiger deaths.csv](Datasets/tiger%20deaths.csv) contains the main activities of the 88 people at the time they were attacked and killed. The header is called `activity`.

1. Import the pandas and seaborn modules.
2. Read in the csv file [Datasets/tiger deaths.csv](Datasets/tiger%20deaths.csv) into a pandas DataFrame called `tigers`. 
3. Make a bar plot of `tigers['activity']`. Consider whether the bars should be vertical or horizontal.
4. Add some useful annotation.

````python
import pandas as pd
import seaborn as sns

tiger = pd.read_csv('Datasets/tiger deaths.csv')

g = sns.displot(y=tiger['activity'], shrink=0.8)

g.ax.set_ylabel('Activity at time of death')
g.ax.set_xlabel('Number of deaths')
g.ax.set_title('Activity of 88 people at time of tiger attack');
````

The Spanish flu killed millions of people around the world between 1918 and 1920. The file [Datasets/spanish flu 1918.csv](Datasets/spanish%20flu%201918.csv) contains the ages of everyone who died in Switzerland in 1918. This includes deaths due to spanish flu as well as the usual causes. The header is called `age`.

1. Import the pandas and seaborn modules.
2. Read in the csv file [Datasets/spanish flu 1918.csv](Datasets/spanish%20flu%201918.csv) into a pandas DataFrame called `flu`. 
3. Plot a histogram of `flu['age']`.
4. Add some useful annotation.

````python
import pandas as pd
import seaborn as sns

flu = pd.read_csv('Datasets/spanish flu 1918.csv')

g = sns.displot(flu['age'])

g.ax.set_xlabel('Age (years)')
g.ax.set_ylabel('Number of deaths')
g.ax.set_title('Age of death in Switzerland in 1918');
````

<div class="alert alert-success">

The large peak at zero corresponds to normal infant mortality. The wide peak around 75 corresponds to normal mortality of the elderly. These two peaks would have been typical in Europe at the beginning of the 20<sup>th</sup> century. 

The wide peak around 25 is due to Spanish flu causing a disproportionate number of deaths in people in their twenties and thirties.

Histograms with more than one peak are called **multi-modal**. Two-peaked histograms are called **bimodal**.

Elephant seals defend harems that may contain hundreds of reproductively active females. Researchers recorded the numbers of females in harems in a population on South Georgia Island. The data are in the file [Datasets/elephant seal.csv](Datasets/elephant%20seal.csv). The header is called `Harem size`.

1. Import the pandas module.
2. Read in the csv file [Datasets/elephant seal.csv](Datasets/elephant%20seal.csv) into a pandas DataFrame called `seal`. 
3. Plot a histogram of `seal['Harem size']`.
4. Add some useful annotation.

````python
import pandas as pd
import seaborn as sns

seal = pd.read_csv('Datasets/elephant seal.csv')

g = sns.displot(seal['Harem size'])

g.ax.set_ylabel('Number of harems')
g.ax.set_title('Harem size of elephant seals on South Georgia Island');
````

<div class="alert alert-success">
    
Most harems contain less than one hundred females. A couple contain over 150 females. 
    
A histogram with this shape is called **right-skewed** or **positive-skewed**.
    
Histograms can also be **left-skewed** or **negative-skewed**.
</div>
